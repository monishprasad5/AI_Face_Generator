```python
import cv2
import dlib
import pandas as pd
import os
import numpy as np
import random
```

```python
def findLipsRectangle(image):
    # load the haar cascades face and smile detectors
    face_detector = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
    lips_detector = cv2.CascadeClassifier("haarcascade_lips.xml")
    cv2recta = []

    # detect faces in the grayscale image
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    face_rects = face_detector.detectMultiScale(gray, 1.1, 8)
    #print('Number of detected faces:', len(face_rects))

    # loop over all the faces detected
    for (x,y,w,h) in face_rects:
        # draw a rectangle in a face
        #cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,255),2)
        #cv2.putText(image, "Face", (x, y), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = image[y:y+h, x:x+w]

        # detecting lips within the face roi
        lips = lips_detector.detectMultiScale(roi_gray, 1.8, 20)
        if len(lips) > 0:
            for (sx, sy, sw, sh) in lips:
                #cv2.rectangle(roi_color, (sx, sy), ((sx + sw), (sy + sh)), (0, 0, 255), 2)
                cv2recta = roi_color[sy:sy+sh,sx:sx+sw]
    return cv2recta
```

```python
# load all images in a directory into memory for task2
width = 800
height = 600
def load_images(path, emotion, size=(256,256)):
    data_list = []
    # enumerate filenames in directory, assume all are images
    for root, subdirs, files in os.walk(path):
        for filename in files:
            file_path = os.path.join(root, filename)
            print("MB: filename",filename)
            # load and resize the image
            image = cv2.imread(file_path)
            image = cv2.resize(image, (width, height))
            # Add result text to the image that will be displayed
            cv2rect = findLipsRectangle(image)
            # store
            new_data = [(cv2rect,filename)]
            data_list.append(new_data)
    return data_list
```

```python
#  for task2
path = 'TD_RGB_E_AllSet/1/original/neutral'
# load dataset
l1 = load_images(path, 'neutral')
path = 'TD_RGB_E_AllSet/1/original/smile'
l2 = load_images(path, 'smile')

data = l1

for i in range(0,len(data)):
    if len(data[i][0][0]) > 0:
        img_path = 'TD_RGB_E_AllSet/1/cropped/neutral/'+data[i][0][1]
        cv2.imwrite(img_path,data[i][0][0])
        
data = l2

for i in range(0,len(data)):
    if len(data[i][0][0]) > 0:
        img_path = 'TD_RGB_E_AllSet/1/cropped/smile/'+data[i][0][1]
        cv2.imwrite(img_path,data[i][0][0])
```

```python
# for task1
# load all images in a directory into memory
width = 800
height = 600
def load_images(path, emotion, size=(256,256)):
    data_list = []
    # enumerate filenames in directory, assume all are images
    for root, subdirs, files in os.walk(path):
        for filename in files:
            file_path = os.path.join(root, filename)
            if (emotion == 'neutral' and 'TD_RGB_E_1.jpg' in filename) or (emotion == 'smile' and 'TD_RGB_E_2.jpg' in filename):
                # load and resize the image
                image = cv2.imread(file_path)
                image = cv2.resize(image, (width, height))
                # Add result text to the image that will be displayed
                cv2rect = findLipsRectangle(image)
                # store
                new_data = [(cv2rect,emotion)]
                data_list.append(new_data)
    return data_list
```

```python
path = 'TD_RGB_E_AllSet'
# load dataset
l1 = load_images(path, 'neutral')
l2 = load_images(path, 'smile')

data = l1 + l2

# shuffling data
random.shuffle(data)

os.mkdir(path+"/extracted_lips")
os.mkdir(path+"/extracted_lips/neutral")
os.mkdir(path+"/extracted_lips/smile")
for i in range(0,len(data)):
    if len(data[i][0][0]) > 0:
        img_path = 'TD_RGB_E_AllSet/extracted_lips/'+data[i][0][1]+'/'+data[i][0][1]+'_'+str(i)+'.jpg'
        cv2.imwrite(img_path,data[i][0][0])
```

```python
# load all celeba images in a directory into memory
width = 800
height = 600
def load_celeba_images(path, lines, size=(256,256)):
    data_list = []
    # enumerate filenames in directory, assume all are images
    for root, subdirs, files in os.walk(path):
        for filename in files[30002:45002]:
            file_path = os.path.join(root, filename)
            print("MB: filename",filename)
            # load and resize the image
            image = cv2.imread(file_path)
            image = cv2.resize(image, (width, height))
            # Add result text to the image that will be displayed
            lineNumInTxt = int(filename.split('.')[0])
            emotionId = int(lines[lineNumInTxt+1].split(' ')[32])
            if emotionId == 1:
                emotion = 'smile'
            else:
                emotion = 'neutral'
            cv2rect = findLipsRectangle(image)
            # store
            new_data = [(cv2rect,emotion)]
            data_list.append(new_data)
    return data_list
```

```python
path = 'data\celeba'
with open(path+'\list_attr_celeba.txt') as f:
    lines = f.readlines()

# load dataset
l3 = load_celeba_images(path+'\images', lines)

for i in range(0,len(l3)):
    if len(l3[i][0][0]) > 0:
        img_path = 'TD_RGB_E_AllSet/extracted_lips/'+l3[i][0][1]+'/'+l3[i][0][1]+'_celeba_'+str(i)+'.jpg'
        cv2.imwrite(img_path,l3[i][0][0])
```
